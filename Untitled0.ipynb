```python
%%manim -qm -v WARNING SquareToCircle

class SquareToCircle(Scene):
   def construct(self):
      square = Square()
      circle = Circle()
      circle.set_fill(PINK, opacity=0.5)
      self.play(Create(square))
      self.play(Transform(square, circle))
      self.wait()
```

```python
%%manim -qm -v WARNING HelloWorld

class HelloWorld(Scene):
    def construct(self):
        text = Text("hello!", font_size=100)
        image = ImageMobject("paimeng2.png").scale(0.5)
        image.generate_target()
        image.target.scale(0.5).shift(LEFT * 5.5 + DOWN * 2.5)
        image.shift(2 * LEFT)
        text.shift(1.5 * RIGHT)
        self.play(FadeIn(image))
        self.add(text)
        #self.wait(1)
        self.play(FadeOut(text))
        self.play(MoveToTarget(image))
        #self.wait(1)
        text1 = Text("",font_size=50)
        self.add(text1)
        self.wait(5)
        self.play(FadeOut(text1))
        text2=Text("2010-2019/",font_size=20)

        data_points = [(2010,257),(2011,262),(2012,255),(2013,294),(2014,296),(2015,298),(2016,309),(2017,312),(2018,322),(2019,323),(2020,336)]
        data_points1 = [(2010,254),(2011,257),(2012,252),(2013,259),(2014,262),(2015,265),(2016,265),(2017,268),(2018,275),(2019,278),(2020,291)]
        # 
        axes = Axes(
            x_range=[2009,2020,1],
            y_range=[0,350,50],
            axis_config={"color": BLUE},
            x_length=8,
            y_length=5,
            x_axis_config={"font_size":25},
            y_axis_config={"font_size":25},
        )
        axes.shift(0.5*DOWN)
        axes.add_coordinates()
        legend = VGroup(
            Line(ORIGIN+2*RIGHT, ORIGIN+3*RIGHT, color=RED),
            Text("", color=RED,font_size=20).next_to(ORIGIN+3*RIGHT, RIGHT)
        )
        legend1 = VGroup(
            Line(ORIGIN+2*RIGHT+DOWN, ORIGIN+3*RIGHT+DOWN, color=YELLOW),
            Text("", color=RED,font_size=20).next_to(ORIGIN+3*RIGHT+DOWN, RIGHT)
        )
        text2.to_edge(UP)
        self.play(FadeIn(text2))
        self.play(Create(axes))
        label=axes.get_axis_labels()
        self.add(label)
        self.play(Create(legend),Create(legend1))
        # 
        line_graph = VMobject()
        value_labels = VGroup()
        dot_group = VGroup()

        for i in range(len(data_points) - 1):
            start_point = axes.coords_to_point(*data_points[i])
            end_point = axes.coords_to_point(*data_points[i + 1])
            line = Line(start_point, end_point, color=RED)
            line_graph.add(line)
            # 
            dot = Dot(color=RED).move_to(start_point)
            dot_group.add(dot)
            # 
            value_label = Text(f"{data_points[i][1]}", color=RED,font_size=20)
            value_label.next_to(start_point, UP)
            value_labels.add(value_label)
        line_graph1 = VMobject()
        value_labels1 = VGroup()
        dot_group1 = VGroup()
        for i in range(len(data_points1) - 1):
            start_point1 = axes.coords_to_point(*data_points1[i])
            end_point1 = axes.coords_to_point(*data_points1[i + 1])
            line1 = Line(start_point1, end_point1, color=YELLOW)
            line_graph1.add(line1)
            # 
            dot = Dot(color=YELLOW).move_to(start_point1)
            dot_group1.add(dot)
            # 
            value_label1 = Text(f"{data_points1[i][1]}", color=YELLOW,font_size=20)
            value_label1.next_to(start_point1, DOWN)
            value_labels1.add(value_label1)


        self.play(AnimationGroup(*[Create(dot) for dot in dot_group], lag_ratio=0),AnimationGroup(*[Create(dot) for dot in dot_group1], lag_ratio=0))
        self.play(AnimationGroup(*[Create(value_label) for value_label in value_labels], lag_ratio=0),AnimationGroup(*[Create(value_label1) for value_label1 in value_labels1], lag_ratio=0))
        # 
        self.play(Create(line_graph),Create(line_graph1))
        #self.wait(1)
        self.play(FadeOut(label),FadeOut(axes),FadeOut(line_graph),FadeOut(line_graph1),FadeOut(dot_group),FadeOut(dot_group1),FadeOut(legend),FadeOut(legend1),FadeOut(value_labels),FadeOut(value_labels1))
        #self.wait(1)
        text3=Text("",color=RED,font_size=40)
        text3.shift(2.5*UP)

        # text4=Text("",color=WHITE,font_size=40)
        # text4.next_to(text3.get_left(),DOWN*4+RIGHT)
        # text5 = Text("",color=WHITE,font_size=40)
        # text4.next_to(text4.get_left(),DOWN*2+RIGHT)
        script = '''
        

        

        

        
        '''
        image1 = ImageMobject("paimeng1.png").scale(0.5)
        image1.scale(0.5).shift(LEFT * 5.5 + DOWN * 2.5)
        text4 = Text(script,font_size=40)
        text4.next_to(text3.get_left(),DOWN*4+RIGHT)
        self.play(Transform(text2,text3),FadeOut(image),FadeIn(image1))
        self.play(Write(text4))
        self.wait(0.5)
        self.play(FadeOut(text4),FadeOut(text2))
        #self.wait()
        text5=Text("",font_size=40)
        text5.shift(LEFT*3+2*UP*1.5)
        text6=Text("",font_size=40)
        text6.shift(RIGHT*3+2*UP*1.5)
        self.play(FadeIn(text5),FadeIn(text6))
        script1 = '''
        

        

        
        '''
        script2 = '''
        !

        !

        !
        '''
        text7=Text(script1,font_size=30)
        text7.shift(LEFT*3)
        text8=Text(script2,font_size=30,color=RED)
        text8.shift(RIGHT*3)
        self.play(FadeIn(text7),FadeIn(text8))
        self.wait(6)
        self.play(FadeOut(text5),FadeOut(text6),FadeOut(text7),FadeOut(text8),FadeOut(image1))
        image2 = ImageMobject("paimeng3.png").scale(3)
        image2.scale(0.5).shift(LEFT*2)

        script3 = '''
        
        
        
        
        '''
        text9=Text(script3,font_size=40)
        text9.shift(RIGHT*1.5)
        self.play(SpinInFromNothing(image2),FadeIn(text9))
        self.play(Flash(image2, color=YELLOW, flash_radius=0.5))
        self.wait(1)
        self.play(FadeOut(image2),FadeOut(text9),SpinInFromNothing(image))
        image3 = ImageMobject("dsa.png")
        image3.scale(0.5).shift(LEFT*3+UP).scale(2)
        image4 = ImageMobject("ct.png").scale(2.5)
        image4.scale(0.5).shift(RIGHT*3+0.2*UP)
        self.play(FadeIn(image3),FadeIn(image4))
        text10=Text("DSA",font_size=20)
        text10.next_to(image3,DOWN)
        text11=Text("CT",font_size=20)
        text11.next_to(image4,DOWN)
        self.play(FadeIn(text10),FadeIn(text11))
        self.wait(15)
        self.play(FadeOut(image3),FadeOut(image4),Uncreate(text10),Uncreate(text11))
        image5 = ImageMobject("dl1.png",invert=True).scale(2.5)
        image5.scale(0.5).shift(UP)
        text12=Text("",font_size=20)
        text12.next_to(image5,DOWN)
        self.play(FadeIn(image5),FadeIn(text12))
        self.wait(1)
        self.play(FadeOut(image5),Uncreate(text12))
        image6 = ImageMobject("net1.png",invert=True).scale(4)
        image6.scale(0.5).shift(UP)
        text13=Text("U-Net",font_size=20)
        text13.next_to(image6,DOWN)
        self.play(FadeIn(image6),FadeIn(text13))
        self.wait(5)
        self.play(FadeOut(image6),Uncreate(text13))
        image7 = ImageMobject("net2.png",invert=True).scale(4)
        image7.scale(0.5).shift(UP)
        text14=Text("VoxelMorph",font_size=20)
        text14.next_to(image7,DOWN)
        self.play(FadeIn(image7),FadeIn(text14))
        self.wait(5)
        self.play(FadeOut(image7),Uncreate(text14),FadeOut(image))
        self.play(FadeIn(image2))
        self.wait(4.5)
        self.play(FadeOut(image2),SpinInFromNothing(image))
        image8 = ImageMobject("ds.png")
        image8.scale(0.5).shift(LEFT*3+0.2*UP).scale(2)
        image9 = ImageMobject("cta.png").scale(2.5)
        image9.scale(0.5).shift(RIGHT*3+0.25*UP)
        text15=Text("DSA",font_size=20)
        text15.next_to(image8,DOWN,buff=0.77)
        text16=Text("CTA",font_size=20)
        text16.next_to(image9,DOWN,buff=0.5)
        self.play(GrowFromEdge(image8,LEFT),GrowFromEdge(image9,LEFT),Create(text15),Create(text16))
        self.wait(6)
        self.play(FadeOut(image8),FadeOut(image9),FadeOut(text15),FadeOut(text16))
        image10 = ImageMobject("peizhun.png",invert=True).scale(4)
        image10.scale(0.5)
        text17=Text("",font_size=20)
        text17.next_to(image10,DOWN)
        self.play(FadeIn(image10),FadeIn(text17))
        self.wait(4)
        self.play(FadeOut(image10),FadeOut(text17))
        image11 = ImageMobject("jilian.png",invert=True).scale(4)
        image11.scale(0.5)
        text18=Text("VTN",font_size=20)
        text18.next_to(image11,DOWN)
        self.play(FadeIn(image11),FadeIn(text18))
        self.wait(6.5)
        #self.play(FadeOut(image11),FadeOut(text18))
        text19=Text("",font_size=50,color=RED).shift(LEFT+2*UP)
        text20=Text("",font_size=60,color='#0418CF').next_to(text19,RIGHT)
        self.play(FadeTransform(image11,text19),FadeTransform(text18,text20))
        self.wait(1)
        text21=Text("",font_size=40,color='#0418CF').shift(LEFT*3)
        text22=Text("",font_size=40,color='#FF006F')
        text23=Text("",font_size=40,color='#FF6D47').shift(RIGHT*3)
        self.play(DrawBorderThenFill(text21),DrawBorderThenFill(text22),DrawBorderThenFill(text23))
        self.wait(22.5)
        image.target.scale(2).shift(RIGHT*5.5+UP*3)
        self.play(FadeOut(text21,shift=LEFT),FadeOut(text22,shift=DOWN),FadeOut(text23,shift=RIGHT),FadeOut(text20,shift=UP),FadeOut(text19,shift=UP))
        self.play(MoveToTarget(image))
        self.wait(1)
```
